Deleting the countries that have no values at all for one variable:

```python
def clean_all_na(df, key, var):
    grouped = df.groupby(key, as_index = False).mean()
    grouped_na = grouped[grouped[var].isnull()]
    df = df[~df[key].isin(grouped_na[key])]
    return df
```

```python
for x in list(digital.columns)[2:]:
    digital = clean_all_na(digital, 'country', x)
```
